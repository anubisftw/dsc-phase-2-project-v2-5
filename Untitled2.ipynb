```python
# split it
X = df[cols]
y = df['price']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .20 , random_state = 69)
```

```python
ohe = OneHotEncoder(handle_unknown = 'ignore')

ohe_cols2 = ['zip']
ohe_cols = ['zip', 'heat_source', 'sewer_system', 'city']

ohe.fit(X_train[ohe_cols])
ohe_train = ohe.transform(X_train[ohe_cols]).toarray()
ohe_test = ohe.transform(X_test[ohe_cols]).toarray()

ohe_train = pd.DataFrame(ohe_train, columns = ohe.get_feature_names())
ohe_test = pd.DataFrame(ohe_test, columns = ohe.get_feature_names())

ohe_train.index = X_train.index
ohe_test.index = X_test.index 

X_train = pd.concat([X_train, ohe_train],1)
X_test = pd.concat([X_test, ohe_test],1)

X_train.drop(ohe_cols,1,inplace =True)
X_test.drop(ohe_cols,1, inplace = True) 

```

```python
#check amount of rows and cols
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()


lr.fit(X_train,y_train)
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
preds = lr.predict(X_test)


```
